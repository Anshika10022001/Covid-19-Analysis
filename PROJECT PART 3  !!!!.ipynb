```python
import requests
import json
from pprint import pprint
import pandas as pd
from datetime import datetime,date
```

```python
response= requests.get("https://data.covid19india.org/v4/min/data.min.json")
results= json.loads(response.text)
```

```python
df= pd.DataFrame(results)
df
```

```python
df.drop('delta',inplace =True)
df.drop('delta21_14',inplace =True)
df.drop('delta7',inplace =True)
df.drop('meta',inplace =True)
df.drop('districts',inplace =True)
```

```python
df2=df.transpose()
df2
```

```python
new=df2['total'].apply(pd.Series)
new
```

```python
new.reset_index(level=0,inplace=True)
```

```python
new.rename({'index':'state_codes'},axis=1,inplace=True)
new
```
