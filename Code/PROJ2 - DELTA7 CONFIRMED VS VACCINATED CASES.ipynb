# DELTA7 CONFIRMED CASES VS VACCINATED CASES 

### The main aim of second part of analysis is to Compare delta7 vaccinated cases with confirmed cases

```python
#importing important libraries
import requests
import json
from pprint import pprint
import pandas as pd
from datetime import datetime,date
```

```python
#Using the requests library to send an HTTP GET request to the specified URL. 
#The get() function retrieves the content of the URL and returns a Response object.
response= requests.get("https://data.covid19india.org/v4/min/data.min.json")

#Extracts the response content as text and uses the json.loads() function to parse the JSON data into a Python object.
#The parsed data is stored in the results variable.
results= json.loads(response.text)
```

```python
#Printing the type of results
type(results)
```

```python
#This DataFrame represents the JSON data in a tabular format, where each key-value pair in the JSON becomes a column in the DataFrame.
df= pd.DataFrame(results)
df
```

```python
#Dropping the irrelevant rows from df as our main focus is only on delta7 row
df.drop('delta',inplace =True)
df.drop('delta21_14',inplace =True)
df.drop('districts',inplace =True)
df.drop('meta',inplace =True)
df.drop('total',inplace =True)
```

```python
#Transposing the dataframe df for better visual
df2=df.transpose()
df2
```

```python
#creating a new DataFrame called DELTA by extracting the 'delta7' column from the DataFrame df2
#Also expanding it into separate columns using the apply() function with pd.Series.
DELTA=df2['delta7'].apply(pd.Series)
DELTA
```

```python
#modifying the DataFrame DElta by resetting the index and making the current index values become a column in the DataFrame.
DELTA.reset_index(level=0,inplace=True)
#modifying the DataFrame AN by renaming the column labeled 'index' to 'state_codes'
DELTA.rename({'index':'state_codes'},axis=1,inplace=True)
DELTA
```
